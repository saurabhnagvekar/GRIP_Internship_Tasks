### Visualizing the Data

```python
for i in iris.select_dtypes(include=(np.number)).columns[1:]:
    plt.figure(figsize=(20,8))
    iris[i].plot(kind="kde")
    plt.title("Distribution plot",fontsize=15)
    plt.xlabel(i,fontsize=15)
    plt.show()
    print('--'*62)
```

```python
for i in iris.select_dtypes(include=(np.number)).columns[1:]:
    plt.figure(figsize=(20,8))
    sns.boxplot(y = iris[i])
    plt.title("Box plot",fontsize=15)
    plt.xlabel(i,fontsize=15)
    plt.show()
    print('--'*62)
```

### Preparing the Data Model:

```python
X = iris.iloc[:,1:-1]
y = iris.iloc[:,-1]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
print("X_train shape:",X_train.shape)
print("X_test shape:",X_test.shape)
print("y_train shape:",y_train.shape)
print("y_test shape:",y_test.shape)
```

### Building Decision Tree Model:

```python
DT = DecisionTreeClassifier()
```

```python
DT_model = DT.fit(X_train,y_train)
```

### Visualising Decision Tree:

```python
from sklearn import tree

plt.figure(figsize=(20,10))
fig = tree.plot_tree(DT_model,feature_names = X_train.columns, class_names=["Iris-setosa","Iris-virginica","Iris-versicolor"],
                     filled=True,rounded=True)
```
